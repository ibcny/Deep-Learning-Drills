```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
%matplotlib inline  
```

```python
mnist = input_data.read_data_sets('data/', one_hot=True)
trainimg   = mnist.train.images
trainlabel = mnist.train.labels
testimg    = mnist.test.images
testlabel  = mnist.test.labels
```

```python
n_input_width       = 28
n_input_height      = 28
n_input_channel     = 1

n_conv1_patch_size  = 3
n_conv1_filter      = 32
n_conv2_patch_size  = 3
n_conv2_filter      = 64
n_conv3_patch_size  = 3
n_conv3_filter      = 128

# tf Graph input
x = tf.placeholder(tf.float32, [None, n_input_width * n_input_height])
y = tf.placeholder(tf.float32, [None, n_input_width * n_input_height])

wc1 = tf.Variable(tf.random_normal([n_conv1_patch_size, n_conv1_patch_size, n_input_channel, n_conv1_filter], stddev=0.1))
bc1 = tf.Variable(tf.random_normal([n_conv1_filter], stddev=0.1))

wc2 = tf.Variable(tf.random_normal([n_conv2_patch_size, n_conv2_patch_size, n_conv1_filter, n_conv2_filter], stddev=0.1))
bc2 = tf.Variable(tf.random_normal([n_conv2_filter], stddev=0.1))

wc3 = tf.Variable(tf.random_normal([n_conv3_patch_size, n_conv3_patch_size, n_conv2_filter, n_conv3_filter], stddev=0.1))
bc3 = tf.Variable(tf.random_normal([n_conv3_filter], stddev=0.1))

# Reshape input
input_r = tf.reshape(x, shape=[-1, n_input_width, n_input_width, n_input_channel])
# Convolution
conv = tf.nn.conv2d(input_r, wc1, strides=[1, 1, 1, 1], padding='SAME')
# Add-bias
bias_1 = tf.nn.bias_add(conv, bc1) # broadcasting is supported
# Pass ReLu
relu_1 = tf.nn.relu(bias_1)
# Max-pooling
pool  = tf.nn.max_pool(relu_1, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='SAME')

# Convolution
conv_2 = tf.nn.conv2d(pool, wc2, strides=[1, 1, 1, 1], padding='SAME')
# Add-bias
bias_2 = tf.nn.bias_add(conv_2, bc2) # broadcasting is supported
# Pass ReLu
relu_2 = tf.nn.relu(bias_2)
# Max-pooling
pool_2  = tf.nn.max_pool(relu_2, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='SAME')

# Convolution
conv_3 = tf.nn.conv2d(pool_2, wc3, strides=[1, 1, 1, 1], padding='SAME')
# Add-bias
bias_3 = tf.nn.bias_add(conv_3, bc3) # broadcasting is supported
# Pass ReLu
relu_3 = tf.nn.relu(bias_3)
# Max-pooling - it might be ignored
pool_3  = tf.nn.max_pool(relu_3, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='SAME')

####################################################
#  the min layer is a 7x7x64
####################################################
####################################################
# Start upsampling
####################################################

cur_batch_size = tf.shape(x)[0]

# Upsample 3. 
wd3 = tf.Variable(tf.random_normal([n_conv3_patch_size, n_conv3_patch_size, n_conv2_filter , n_conv3_filter], stddev=0.1))
bd3 = tf.Variable(tf.random_normal([n_conv2_filter], stddev=0.1))
d3_out_shape = tf.pack([cur_batch_size, int(n_input_width/4), int(n_input_height/4), n_conv2_filter])
convt_3 = tf.nn.relu(tf.nn.conv2d_transpose(pool_3, filter=wd3, output_shape = d3_out_shape, strides=[1,2,2,1], padding='SAME') + bd3)

# Upsample 2. 
wd2 = tf.Variable(tf.random_normal([n_conv2_patch_size, n_conv2_patch_size, n_conv1_filter , n_conv2_filter], stddev=0.1))
bd2 = tf.Variable(tf.random_normal([n_conv1_filter], stddev=0.1))
d2_out_shape = tf.pack([cur_batch_size, int(n_input_width/2), int(n_input_height/2), n_conv1_filter])
convt_2 = tf.nn.relu(tf.nn.conv2d_transpose(convt_3, filter=wd2, output_shape = d2_out_shape, strides=[1,2,2,1], padding='SAME') + bd2)

# Upsample 1. 
wd1 = tf.Variable(tf.random_normal([n_conv1_patch_size, n_conv1_patch_size, n_input_channel, n_conv1_filter], stddev=0.1))
bd1 = tf.Variable(tf.random_normal([n_input_channel], stddev=0.1))
d1_out_shape = tf.pack([cur_batch_size, int(n_input_width), int(n_input_height), n_input_channel])
convt_1 = tf.nn.relu(tf.nn.conv2d_transpose(convt_2, filter=wd1, output_shape = d1_out_shape, strides=[1,2,2,1], padding='SAME') + bd1)

# add dropout
dropout_keep_prob = tf.placeholder("float")
out = tf.nn.dropout(convt_1, dropout_keep_prob) 
y_conv = tf.reshape(out, [-1, n_input_height * n_input_width])
cost = tf.reduce_mean(tf.pow(y_conv - y, 2))

print("done")
```

```python
# Learning Parameters
learning_rate       = 0.001
batch_size          = 100
plot_step           = 5
display_step        = 1
training_epochs     = 50

optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(cost) # Adam Optimizer
init = tf.initialize_all_variables()

# Saver
savedir = "nets/"
saver = tf.train.Saver(max_to_keep=training_epochs) 

# Launch the graph
sess = tf.Session()

# Training 
sess.run(init)

for epoch in range(training_epochs):
    avg_cost = 0.
    num_batch = int(mnist.train.num_examples/batch_size)
    # Loop over all batches
    for i in range(num_batch): 
        randidx = np.random.randint(trainimg.shape[0], size=batch_size)
        batch_xs = trainimg[randidx, :]
        batch_xs_noisy = batch_xs + 0.3*np.random.randn(batch_xs.shape[0], 784)

        batch_ys = trainlabel[randidx, :]                

        # Fit training using batch data
        sess.run(optimizer, feed_dict={x: batch_xs_noisy, y: batch_xs, dropout_keep_prob: 0.5})
        # Compute average loss
        avg_cost += sess.run(cost, feed_dict={x: batch_xs_noisy, y: batch_xs, dropout_keep_prob: 1})/num_batch

    # Display logs per epoch step
    if epoch % display_step == 0:
        print ("Epoch: %03d/%03d cost: %.9f" % (epoch, training_epochs, avg_cost))
    if epoch % plot_step == 0:
        # Test one
        randidx  = np.random.randint(testimg.shape[0], size=1)
        testvec  = testimg[randidx, :]
        noisyvec = testvec + 0.3*np.random.randn(1, 784)
        outvec   = sess.run(out, feed_dict={x: testvec, dropout_keep_prob: 1.})
        outimg   = np.reshape(outvec, (28, 28))

        # Plot 
        plt.matshow(np.reshape(testvec, (28, 28)), cmap=plt.get_cmap('gray'))
        plt.title("[" + str(epoch) + "] Original Image")
        plt.colorbar()

        plt.matshow(np.reshape(noisyvec, (28, 28)), cmap=plt.get_cmap('gray'))
        plt.title("[" + str(epoch) + "] Input Image")
        plt.colorbar()

        plt.matshow(outimg, cmap=plt.get_cmap('gray'))
        plt.title("[" + str(epoch) + "] Reconstructed Image")
        plt.colorbar()
        plt.show()

        saver.save(sess, savedir + 'dae_dr.ckpt', global_step=epoch)
print ("Optimization Finished!")
```

```python
# Restore
epoch = 45
saver.restore(sess, "nets/dae_dr.ckpt-" + str(epoch))
```

```python
# Test 
randidx   = np.random.randint(testimg.shape[0], size=1)
orgvec    = testimg[randidx, :]
testvec   = testimg[randidx, :]
label     = np.argmax(testlabel[randidx, :], 1)

print ("label is %d" % (label)) 

print ("Gaussian Noise")
noisyvec_gauss = testvec + 0.3*np.random.randn(1, 784)
   
print ("Salt and Pepper Noise")
noisyvec_saltpapper = testvec
rate     = 0.15
noiseidx = np.random.randint(testimg.shape[1], size=int(testimg.shape[1]*rate))
noisyvec_saltpapper[0, noiseidx] = 1-noisyvec[0, noiseidx]

outvec_gauss   = sess.run(out, feed_dict={x: noisyvec_gauss, dropout_keep_prob: 1})
outimg_gauss   = np.reshape(outvec_gauss, (28, 28))

outvec_saltpapper   = sess.run(out, feed_dict={x: noisyvec_saltpapper, dropout_keep_prob: 1})
outimg_saltpapper   = np.reshape(outvec_saltpapper, (28, 28))

# Plot 
plt.matshow(np.reshape(orgvec, (28, 28)), cmap=plt.get_cmap('gray'))
plt.title("Original Image")
plt.colorbar()

plt.matshow(np.reshape(noisyvec_gauss, (28, 28)), cmap=plt.get_cmap('gray'))
plt.title("Input Image with Gaussian noise")
plt.colorbar()

plt.matshow(outimg_gauss, cmap=plt.get_cmap('gray'))
plt.title("Reconstructed Image from Gaussian noise")
plt.colorbar()
plt.show()

plt.matshow(np.reshape(noisyvec_saltpapper, (28, 28)), cmap=plt.get_cmap('gray'))
plt.title("Input Image with salt papper noise")
plt.colorbar()

plt.matshow(outimg_saltpapper, cmap=plt.get_cmap('gray'))
plt.title("Reconstructed Image from salt papper noise")
plt.colorbar()
plt.show()
```
